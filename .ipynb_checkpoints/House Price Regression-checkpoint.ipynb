# Dummy Encoding

```python
len(OHEncoded) + len(LabelEncoded)
```

```python
def create_dummies(categorical_features,OHEncoded):
    all_dummies = pd.DataFrame()
    for feature in OHEncoded:
        dummies = pd.get_dummies(categorical_features[feature], prefix = feature)
        temp = feature + '_None'
        if temp in dummies.columns:
            dummies.drop(temp, axis = 1, inplace = True)
        else:
            dummies.drop(dummies.columns[0], axis = 1, inplace = True)
        all_dummies = pd.concat([all_dummies,dummies], axis = 1)
    return all_dummies
```

```python
all_dummies = create_dummies(categorical_features,OHEncoded)
```

```python
all_dummies
```

```python
categorical_features.drop(OHEncoded, axis = 1, inplace = True)
```

```python
categorical_features = pd.concat([categorical_features, all_dummies], axis = 1)
```
